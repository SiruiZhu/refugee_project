```python
import pandas as pd
```

## Read Refugee.csv

```python
df_refugee = pd.read_csv("/src/data/refugee.csv")
```

```python
df_refugee
```

```python
df_refugee.rename(columns = {'Country / territory of asylum/residence':'country'}, inplace=True)
```

```python
df_refugee
```

```python
df_refugee.shape
```

## delete unknown Orgin countries row

```python
df_refugee = df_refugee[df_refugee['Origin'] != 'Various/Unknown']
```

```python
df_refugee.dtypes
```

```python
df_refugee.shape
```

```python
df_refugee.head()
```

```python
df_track = df_refugee.groupby('Year').country.value_counts().reset_index(name='numbers_origin')
```

```python
df_track
```

```python
df_refugee_track = df_track.groupby('Year').numbers_origin.sum().reset_index(name = 'track_number')
```

```python
df_refugee_track
```

```python
df_refugee_track.to_csv('refugee_track_number.csv',index=False)
```
