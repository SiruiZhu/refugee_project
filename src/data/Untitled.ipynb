### get 2017 refugee data

```python
df_refugee = pd.read_csv("refugee_total_clean.csv")
```

```python
df_refugee.head()
```

```python
df_2017 = df_refugee[df_refugee['Year'] ==2017]
df_2017.head()
```

```python
df_2017 = df_2017[df_2017['value'] != 0]
```

```python
df_2017.to_csv("refugee2017.csv", index=False)
```

```python
df_2017_country = df_2017['country'].append(df_2017['Origin']).reset_index(drop=True).reset_index(name='country')
df_2017_country.head()
```

```python
df_2017_country = df_2017_country.drop_duplicates(subset = 'country', keep='first', inplace=False).reset_index(drop=True)
```

```python
df_2017_country.head()
```

```python
df_2017_country = df_2017_country.drop(['index'],axis=1)
```
